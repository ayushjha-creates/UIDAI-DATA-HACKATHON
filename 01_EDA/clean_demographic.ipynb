```python
import pandas as pd
import re

def clean_state_name(x):
    if pd.isna(x):
        return x
    x = str(x).lower()
    x = re.sub(r'[^a-z\s]', ' ', x)   # remove symbols like &, .
    x = re.sub(r'\s+', ' ', x).strip()  # remove extra spaces
    return x
```

```python
def clean_district_name(x):
    if pd.isna(x):
        return x
    x = str(x).lower()
    x = re.sub(r'[^a-z\s]', ' ', x)   # remove symbols like &, .
    x = re.sub(r'\s+', ' ', x).strip()  # remove extra spaces
    return x
```

```python
df['district'] = df['district'].apply(clean_district_name)
```

```python
# List of demographic variables
demo_vars = ['demo_age_5_17', 'demo_age_17_', 'total_population', 'young_percentage']
```

```python
# Descriptive Statistics
desc_stats = df[demo_vars].describe(percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99]).T
desc_stats['skewness'] = df[demo_vars].skew()
desc_stats['kurtosis'] = df[demo_vars].kurt()

print(desc_stats)
```

```python
# Visualization

# Set up figure for Histograms
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Univariate Distributions of Demographic Variables', fontsize=18)

# Histogram: Age 5-17
sns.histplot(df['demo_age_5_17'], bins=50, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Distribution of Age 5-17 Registrations')
axes[0, 0].set_yscale('log') # Log scale due to likely right skew
axes[0, 0].set_xlabel('Count (Log Scale)')

# Histogram: Age 18+
sns.histplot(df['demo_age_17_'], bins=50, kde=True, ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Distribution of Age 18+ Registrations')
axes[0, 1].set_yscale('log')
axes[0, 1].set_xlabel('Count (Log Scale)')

# Histogram: Total Population
sns.histplot(df['total_population'], bins=50, kde=True, ax=axes[1, 0], color='green')
axes[1, 0].set_title('Distribution of Total Registrations')
axes[1, 0].set_yscale('log')
axes[1, 0].set_xlabel('Count (Log Scale)')

# Histogram: Young Percentage
sns.histplot(df_active['young_percentage'], bins=50, kde=True, ax=axes[1, 1], color='purple')
axes[1, 1].set_title('Distribution of Young Population Percentage (per record)')
axes[1, 1].set_xlabel('Percentage (%)')
```

# Trivariate Aanalysis On Demographic Data Only 

```python
df
```

```python
# specific months where certain states spike
df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')

# Create Missing Columns in Main df

df['month'] = df['date'].dt.month
df_active = df[df['total_population'] > 0].copy()

pivot_vol = df_active.pivot_table(index='state', columns='month', values='total_population', aggfunc='sum')

# Filter top 15 states 
top_states = pivot_vol.sum(axis=1).nlargest(15).index
pivot_vol_top = pivot_vol.loc[top_states]

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_vol_top, cmap='YlGnBu', linewidths=0.5, linecolor='gray')
plt.title('Trivariate: Total Registrations by State and Month')
plt.xlabel('Month (1-12)')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
df['child_ratio_demo'] = (
    df['demo_age_5_17'] / (df['total_demo_updates'] + 1)
)
```

```python
state_summary = (
    df.groupby('state')[['total_demo_updates','child_ratio_demo']]
      .mean()
      .reset_index()
)

plt.figure()
plt.scatter(
    state_summary['total_demo_updates'],
    state_summary['child_ratio_demo']
)

for _, row in state_summary.iterrows():
    plt.text(
        row['total_demo_updates'],
        row['child_ratio_demo'],
        row['state'],
        fontsize=7
    )

plt.xlabel("Average Demographic Updates")
plt.ylabel("Child Update Ratio (5–17)")
plt.title("State-wise Demographic Load vs Child Update Share")
plt.show()
```

```python
# State vs Month vs Total Updates

state_month_total = (
    df.groupby(['state','month'])['total_demo_updates']
      .mean()
      .reset_index()
)

state_to_plot = "Delhi"  # change state

temp = state_month_total[state_month_total['state'] == state_to_plot]

plt.figure()
plt.plot(temp['month'], temp['total_demo_updates'])
plt.xlabel("Month")
plt.ylabel("Average Demographic Updates")
plt.title(f"Seasonal Demographic Update Load – {state_to_plot}")
plt.xticks(range(1,13))
plt.show()
```

```python
# Month vs Age Group vs National Trend

month_age_national = (
    df.groupby('month')[['demo_age_5_17','demo_age_17_']]
      .mean()
)

plt.figure()
plt.plot(month_age_national.index, month_age_national['demo_age_5_17'], label='Age 5–17')
plt.plot(month_age_national.index, month_age_national['demo_age_17_'], label='Age 17+')
plt.xlabel("Month")
plt.ylabel("Average Demographic Updates")
plt.title("National Month-wise Demographic Updates by Age Group")
plt.legend()
plt.xticks(range(1,13))
plt.show()
```

```python
# State Vs Month Vs Spike Detection
state_month_spike = (
    df.groupby(['state','year','month'])['total_demo_updates']
      .sum()
      .reset_index()
)

state_month_spike['rolling_mean'] = (
    state_month_spike.groupby('state')['total_demo_updates']
    .transform(lambda x: x.rolling(3, min_periods=1).mean())
)

state_month_spike['spike_ratio'] = (
    state_month_spike['total_demo_updates'] /
    (state_month_spike['rolling_mean'] + 1)
)

# View extreme month-wise spikes
state_month_spike[state_month_spike['spike_ratio'] > 2].head()
```
