```python
import pandas as pd
import re

def clean_state_name(x):
    if pd.isna(x):
        return x
    x = str(x).lower()
    x = re.sub(r'[^a-z\s]', ' ', x)   # remove symbols like &, .
    x = re.sub(r'\s+', ' ', x).strip()  # remove extra spaces
    return x
```

```python
def clean_district_name(x):
    if pd.isna(x):
        return x
    x = str(x).lower()
    x = re.sub(r'[^a-z\s]', ' ', x)   # remove symbols like &, .
    x = re.sub(r'\s+', ' ', x).strip()  # remove extra spaces
    return x
```

# Univariate Analysis On Biometric Data Only


```python
# Feature Engineering for Analysis

df = df_bio_clean.copy()

df['date'] = pd.to_datetime(df['date'], dayfirst=True)
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

df['total_bio_updates'] = (
    df['bio_age_5_17'] + df['bio_age_17_']
)
# Ratio (handle divide by zero)
df['minor_percentage'] = np.where(df['total_bio_updates'] > 0, 
                                  (df['bio_age_5_17'] / df['total_bio_updates']) * 100, 
                                  0)
```

```python
df
```

```python
# Statistical Analysis
desc_stats = df[['bio_age_5_17', 'bio_age_17_', 'total_bio_updates', 'minor_percentage']].describe(
    percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99]
).T
desc_stats['skew'] = df[['bio_age_5_17', 'bio_age_17_', 'total_bio_updates', 'minor_percentage']].skew()
desc_stats['kurtosis'] = df[['bio_age_5_17', 'bio_age_17_', 'total_bio_updates', 'minor_percentage']].kurt()


print(df.shape)
print(" Descriptive Statistics ")
print(desc_stats)
```

```python
# Sparsity Analysis
zero_counts = (df[['bio_age_5_17', 'bio_age_17_']] == 0).sum()
zero_pct = (zero_counts / len(df)) * 100
print(" Zero Value Analysis:- ")
print(pd.concat([zero_counts, zero_pct], axis=1, keys=['Zero Count', 'Zero %']))
```

```python
# state vs age grp vs month

state_month_age = (
    df.groupby(['state','month'])[['bio_age_5_17','bio_age_17_']]
      .mean()
      .reset_index()
)

state_to_plot = "Madhya Pradesh" 

temp = state_month_age[state_month_age['state'] == state_to_plot]

plt.figure()
plt.plot(temp['month'], temp['bio_age_5_17'], label='Age 5–17')
plt.plot(temp['month'], temp['bio_age_17_'], label='Age 17+')
plt.xlabel("Month")
plt.ylabel("Average Biometric Updates")
plt.title(f"Month-wise Biometric Updates by Age – {state_to_plot}")
plt.legend()
plt.xticks(range(1,13))
plt.show()
```

```python
# STATE vs TOTAL LOAD vs AGE COMPOSITION

state_summary = (
    df.groupby('state')[['bio_age_5_17','bio_age_17_']]
      .sum()
)

state_summary['total'] = (
    state_summary['bio_age_5_17'] + state_summary['bio_age_17_']
)

state_summary['child_share'] = (
    state_summary['bio_age_5_17'] / (state_summary['total'] + 1)
)

plt.figure()
plt.scatter(
    state_summary['total'],
    state_summary['child_share']
)

for s in state_summary.index:
    plt.text(
        state_summary.loc[s,'total'],
        state_summary.loc[s,'child_share'],
        s,
        fontsize=7
    )

plt.xlabel("Total Biometric Updates")
plt.ylabel("Child Share of Biometric Updates")
plt.title("State-wise Biometric Load vs Age Composition")
plt.show()
```

```python
#DISTRICT vs AGE GROUP vs BIOMETRIC CONCENTRATION

district_age = (
    df.groupby('district')[['bio_age_5_17','bio_age_17_']]
      .sum()
)

plt.figure()
plt.scatter(
    district_age['bio_age_5_17'],
    district_age['bio_age_17_'],
    alpha=0.4
)
plt.xlabel("Child Biometric Updates (5–17)")
plt.ylabel("Adult Biometric Updates (17+)")
plt.title("District-level Biometric Concentration by Age")
plt.show()
```

```python
# STATE × TIME × BIOMETRIC SPIKES
state_time = (
    df.groupby(['state','year','month'])['total_bio_updates']
      .sum()
      .reset_index()
)

state_time['rolling_mean'] = (
    state_time.groupby('state')['total_bio_updates']
    .transform(lambda x: x.rolling(3, min_periods=1).mean())
)

state_time['spike_ratio'] = (
    state_time['total_bio_updates'] /
    (state_time['rolling_mean'] + 1)
)

state_time[state_time['spike_ratio'] > 2].head()
```

```python
# MONTH × AGE GROUP × VARIABILITY

month_age_std = (
    df.groupby('month')[['bio_age_5_17','bio_age_17_']]
      .std()
)

plt.figure()
plt.plot(month_age_std.index, month_age_std['bio_age_17_'], label='Adult')
plt.plot(month_age_std.index, month_age_std['bio_age_5_17'], label='Child')
plt.xlabel("Month")
plt.ylabel("Standard Deviation")
plt.title("Month-wise Variability in Biometric Updates by Age")
plt.legend()
plt.xticks(range(1,13))
plt.show()
```
