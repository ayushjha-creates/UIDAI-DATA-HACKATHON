```python
import pandas as pd
import re

def clean_state_name(x):
    if pd.isna(x):
        return x
    x = str(x).lower()
    x = re.sub(r'[^a-z\s]', ' ', x)   # remove symbols like &, .
    x = re.sub(r'\s+', ' ', x).strip()  # remove extra spaces
    return x
```

BIVARIATE Analaysis For Enrolment Dataset only 

```python
import pandas as pd
import matplotlib.pyplot as plt

df = df_enrolment_clean.copy()

df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

df['total_enrolments'] = (
    df['age_0_5'] +
    df['age_5_17'] +
    df['age_18_greater']
)
```

```python
time_enrol = (
    df.groupby(['year', 'month'])['total_enrolments']
      .sum()
      .reset_index()
)

plt.figure()
plt.plot(time_enrol.index, time_enrol['total_enrolments'])
plt.xlabel("Time (Year–Month Index)")
plt.ylabel("Total Enrolments")
plt.title("Aadhaar Enrolments Over Time")
plt.show()
```

```python
age_totals = df[['age_0_5','age_5_17','age_18_greater']].sum()

plt.figure()
age_totals.plot(kind='bar')
plt.xlabel("Age Group")
plt.ylabel("Total Enrolments")
plt.title("Enrolments by Age Group")
plt.show()
```

```python
state_age = (
    df.groupby('state')[['age_0_5','age_5_17','age_18_greater']]
      .sum()
)

state_age_ratio = state_age.div(state_age.sum(axis=1), axis=0)

plt.figure()
state_age_ratio.head(10).plot(kind='bar', stacked=True)
plt.xlabel("State")
plt.ylabel("Proportion")
plt.title("Age-wise Enrolment Composition (Top 10 States)")
plt.show()
plt.savefig('bivariate_state_age_composition.png')
```

```python
district_enrol = df.groupby('district')['total_enrolments'].sum()

plt.figure()
plt.hist(district_enrol, bins=40)
plt.xlabel("Total Enrolments per District")
plt.ylabel("Number of Districts")
plt.title("Distribution of Enrolments Across Districts")
plt.show()
```

```python
df['child_ratio'] = df['age_5_17'] / (df['total_enrolments'] + 1)

plt.figure()
plt.scatter(df['child_ratio'], df['total_enrolments'], alpha=0.3)
plt.xlabel("Child Enrolment Ratio (5–17)")
plt.ylabel("Total Enrolments")
plt.title("Child Enrolment Ratio vs Total Enrolments")
plt.show()
```

```python
# 2. Time vs Age Group Trends (Line Chart)
# Aggregate by Month
df['month_num'] = df['date'].dt.month
monthly_age = df.groupby('month_num')[['age_0_5', 'age_5_17', 'age_18_greater']].sum()

plt.figure(figsize=(12, 6))
plt.plot(monthly_age.index, monthly_age['age_0_5'], marker='o', label='Age 0-5')
plt.plot(monthly_age.index, monthly_age['age_5_17'], marker='s', label='Age 5-17')
plt.plot(monthly_age.index, monthly_age['age_18_greater'], marker='^', label='Age 18+')
plt.title('Bivariate: Seasonal Trends by Age Group')
plt.xlabel('Month')
plt.ylabel('Total Enrolments')
plt.legend()
plt.grid(True)
plt.savefig('bivariate_seasonal_age.png')
```

```python
# 3. State vs Operational Efficiency (Box Plot)
# Efficiency = Daily Enrolment per Pincode
# We use boxplot to see the spread (Are all centers good, or just a few outliers?)
df['total_enrolment'] = df['age_0_5'] + df['age_5_17'] + df['age_18_greater']

top_5_states = df.groupby('state')['total_enrolment'].sum().nlargest(5).index
df_top5 = df[df['state'].isin(top_5_states)]

plt.figure(figsize=(12, 6))
sns.boxplot(x='state', y='total_enrolment', data=df_top5, palette='Set2')
plt.yscale('log') # Log scale because of outliers
plt.title('Bivariate: Operational Spread (Efficiency) of Top 5 States')
plt.ylabel('Daily Enrolment (Log Scale)')
plt.savefig('bivariate_state_efficiency_boxplot.png')
```

```python
# 4. Correlation Matrix (Numerical vs Numerical)
# Is there a correlation between Child Enrolment and Adult Updates at the same center?
# i.e., Do parents update their Aadhaar when enrolling kids?
corr_matrix = df[['age_0_5', 'age_5_17', 'age_18_greater', 'total_enrolment']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Bivariate: Correlation Matrix of Enrolment Types')
plt.savefig('bivariate_correlation.png')

print("Bivariate Analysis Complete.")
print("Correlation Matrix:\n", corr_matrix)
```

Trivariate Analysis For Enrolment Dataset Only

```python
top_10_states = df.groupby('state')['total_enrolment'].sum().nlargest(10).index
heatmap_data = df[df['state'].isin(top_10_states)].groupby(['state', 'month_num'])['total_enrolment'].sum().unstack()

plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='g', linewidths=.5)
plt.title('Trivariate: Monthly Enrolment Intensity by Top 10 States')
plt.xlabel('Month')
plt.ylabel('State')
plt.savefig('trivariate_heatmap_state_month.png')
```

```python
# 2. State x Age Group Composition (Stacked Bar Chart for Top 10 States)
# We calculate the sum of each age group for the top 10 states
state_age_comp = df[df['state'].isin(top_10_states)].groupby('state')[['age_0_5', 'age_5_17', 'age_18_greater']].sum()
# Normalize to percentage to see the composition structure
state_age_comp_pct = state_age_comp.div(state_age_comp.sum(axis=1), axis=0) * 100

state_age_comp_pct.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='viridis')
plt.title('Trivariate: Age Demographic Structure of Enrolments in Top 10 States')
plt.ylabel('Percentage of Enrolments')
plt.xlabel('State')
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('trivariate_state_age_structure.png')
```

```python
# district vs age vs enrolment concentration
district_age = (
    df_enrolment_clean
    .groupby('district')[['age_0_5', 'age_5_17', 'age_18_greater']]
    .sum()
)

plt.figure()
plt.boxplot(
    [
        district_age['age_0_5'],
        district_age['age_5_17'],
        district_age['age_18_greater']
    ],
    labels=['0–5', '5–17', '18+']
)

plt.ylabel("Enrolments per District")
plt.title("District-wise Enrolment Distribution by Age Group")
plt.show()
```

```python
# How does Age 0-5 vs 5-17 trend differ in diffn states?
top_5_states = top_10_states[:5]
df_melted = df[df['state'].isin(top_5_states)].melt(
    id_vars=['state', 'month_num'], 
    value_vars=['age_0_5', 'age_5_17'], # Focusing on child enrolment
    var_name='Age_Group', 
    value_name='Enrolments'
)
# Aggregate for cleaner lines
df_melted_agg = df_melted.groupby(['state', 'month_num', 'Age_Group'])['Enrolments'].sum().reset_index()

g = sns.FacetGrid(df_melted_agg, col="state", hue="Age_Group", col_wrap=3, height=4, sharey=False)
g.map(sns.lineplot, "month_num", "Enrolments", marker="o")
g.add_legend()
g.fig.suptitle('Trivariate: Child Enrolment Trends (0-5 vs 5-17) across Top 5 States', y=1.05)
plt.savefig('trivariate_facet_trends.png')
```
