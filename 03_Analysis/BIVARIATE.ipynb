# Bivariate Analysis

```python
# total enrolments VS total updates

plt.figure()
plt.scatter(
    merged['total_enrolments'],
    merged['total_updates'],
    alpha=0.3
)
plt.xlabel("Total Enrolments")
plt.ylabel("Total Updates")
plt.title("Total Enrolments vs Total Updates")
plt.show()
```

```python
# total emrolments VS friction index

plt.figure()
plt.scatter(
    merged['total_enrolments'],
    merged['digital_friction_index'],
    alpha=0.3
)
plt.xlabel("Total Enrolments")
plt.ylabel("Digital Friction Index")
plt.title("Enrolments vs Digital Friction")
plt.show()
```

```python
# total updates VS friction index

plt.figure()
plt.scatter(
    merged['total_updates'],
    merged['digital_friction_index'],
    alpha=0.3
)
plt.xlabel("Total Updates")
plt.ylabel("Digital Friction Index")
plt.title("Update Load vs Digital Friction")
plt.show()
```

```python
# Adult System Stress vs Digital Friction Index

plt.figure()
plt.scatter(
    merged['adult_system_stress'],
    merged['digital_friction_index'],
    alpha=0.3
)
plt.xlabel("Adult System Stress")
plt.ylabel("Digital Friction Index")
plt.title("Adult Stress vs Digital Friction")
plt.show()
```

```python
#Child System Stress vs Digital Friction Index

plt.figure()
plt.scatter(
    merged['child_system_stress'],
    merged['digital_friction_index'],
    alpha=0.3
)
plt.xlabel("Child System Stress")
plt.ylabel("Digital Friction Index")
plt.title("Child Stress vs Digital Friction")
plt.show()
```

```python
# Month vs Digital Friction Index

month_friction = (
    merged.groupby('month')['digital_friction_index']
    .mean()
)

plt.figure()
plt.plot(month_friction.index, month_friction.values, marker='o')
plt.xlabel("Month")
plt.ylabel("Average Digital Friction Index")
plt.title("Month-wise Digital Friction")
plt.xticks(range(0,13))
plt.show()
```

```python
# State vs Digital Friction Index

state_friction = (
    merged.groupby('state')['digital_friction_index']
    .mean()
    .sort_values(ascending=False)
)

plt.figure()
state_friction.head(15).plot(kind='bar')
plt.xlabel("State")
plt.ylabel("Average Digital Friction Index")
plt.title("Top States by Digital Friction")
plt.show()
```

```python
# District vs Digital Friction Index

district_friction = (
    merged.groupby('district')['digital_friction_index']
    .mean()
)

plt.figure()
plt.hist(district_friction, bins=50)
plt.xlabel("Average Digital Friction (District)")
plt.ylabel("Number of Districts")
plt.title("District-wise Digital Friction Distribution")
plt.show()
```
