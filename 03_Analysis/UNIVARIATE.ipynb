# UNIVARIATE ANALYSIS

```python
# distribution of total enrolments 

plt.figure()
plt.hist(merged['total_enrolments'], bins=50)
plt.xlabel("Total Enrolments")
plt.ylabel("Frequency")
plt.title("Distribution of Total Enrolments")
plt.show()
```

```python
# distribution of total updates 

plt.figure()
plt.hist(merged['total_updates'], bins=50)
plt.xlabel("Total Updates")
plt.ylabel("Frequency")
plt.title("Distribution of Total Updates")
plt.show()
```

```python
# adult  system stress distribution

plt.figure()
plt.hist(merged['adult_system_stress'], bins=50)
plt.xlabel("Adult System Stress")
plt.ylabel("Frequency")
plt.title("Distribution of Adult System Stress")
plt.show()
```

```python
# child system stress distribution

plt.figure()
plt.hist(merged['child_system_stress'], bins=50)
plt.xlabel("Child System Stress")
plt.ylabel("Frequency")
plt.title("Distribution of Child System Stress")
plt.show()
```

```python
# digital friction index

plt.figure()
plt.hist(merged['digital_friction_index'], bins=50)
plt.xlabel("Digital Friction Index")
plt.ylabel("Frequency")
plt.title("Distribution of Digital Friction Index")
plt.show()
```

```python
#Log-Transformed Digital Friction

plt.figure()
plt.hist(merged['log_digital_friction'], bins=50)
plt.xlabel("Log Digital Friction Index")
plt.ylabel("Frequency")
plt.title("Log-Scaled Digital Friction Distribution")
plt.show()
```

```python
#Zero-Enrolment Frequency

zero_enrol_rate = (merged['total_enrolments'] == 0).mean()
zero_enrol_rate
```

```python
#Zero-update Frequency

zero_update_rate = (merged['total_updates'] == 0).mean()
zero_update_rate
```

```python
# outlier share 

merged['friction_outlier'].value_counts(normalize=True)
```

```python
# temporal coverage

merged['month'].value_counts().sort_index().plot(kind='bar')
plt.xlabel("Month")
plt.ylabel("Record Count")
plt.title("Month-wise Record Distribution")
plt.show()
```

```python
# state distribution

merged['state'].value_counts().head(15).plot(kind='bar')
plt.xlabel("State")
plt.ylabel("Record Count")
plt.title("Top States by Record Volume")
plt.show()
```
